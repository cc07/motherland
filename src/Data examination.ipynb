```python
stockList = {}
for symbol in symbolList:
    
    fname = './equity/usa/daily/{}.zip'.format(symbol)
    if not os.path.isfile(fname):
        print('Missing: {}'.format(symbol))
        continue
        
    zf = z(fname)
    df = pd.read_csv(zf.open('{}.csv'.format(symbol).lower()), header=None, names=['Date', 'Open', 'High', 'Low', 'Close', 'Volume'])
    
    if df.shape[0] > 1000:
        stockList[symbol] = df
#     print('{} {}'.format(symbol, stockList[symbol].shape))
```

```python
abnormalDailyStockList = {}

for symbol in stockList:
    stockList[symbol]['Prev_Close'] = stockList[symbol]['Close'].shift(1)
    stockList[symbol]['Return'] = (stockList[symbol]['Close'] - stockList[symbol]['Prev_Close']) / stockList[symbol]['Prev_Close'] * 100
    
    count = len(stockList[symbol][stockList[symbol]['Return'] > 10])
    if count > 5:
        abnormalDailyStockList[symbol] = stockList[symbol]
#         print('{} {}'.format(symbol, count))
```

```python
abnormalWeeklyStockList = {}

for symbol in stockList:
    stockList[symbol]['Prev_5_Close'] = stockList[symbol]['Close'].shift(5)
    stockList[symbol]['Return_5'] = (stockList[symbol]['Close'] - stockList[symbol]['Prev_5_Close']) / stockList[symbol]['Prev_5_Close'] * 100
    
    count = len(stockList[symbol][stockList[symbol]['Return_5'] > 30])
    if count > 5:
        print(symbol)
        abnormalWeeklyStockList[symbol] = stockList[symbol]
```
